```python
from bisect import bisect_right, bisect_left
```

```python
with open('algo1-programming_prob-2sum.txt') as handle:
    data = handle.readlines()
    
data = [int(x) for x in data]
data = sorted(data)
```

```python
data
```

```python
def sum2(array, target):
    Q = set()
    for x in array:
        if x in Q:
            return 1
        else:
            Q.add(target-x)
    return 0
```

```python
count = 0
for t in range(-10000,10001):
    count += sum2(data, t)
    if t+10000%100==0:
        print(t+10000)
```

```python
arr2 = [-3,-1,1,2,9,11,7,6,2]
count = 0
for t in range(3,11):
    count += sum2(arr2, t)
```

```python
pairs = set()
for x in data:
    if x>5000:
        break
    else:
        low = bisect_left(data, -10000-x)
        high = bisect_right(data, 10000-x)
        for k in data[low:high+1]:
            if x + k >=-10000 and x + k <= 10000:
                pairs.add(x+k)
```
